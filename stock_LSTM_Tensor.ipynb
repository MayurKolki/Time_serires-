```python
import pandas as pd
import numpy as np
```

```python
#import librarries
import tensorflow as tf
import keras
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
#read train and test
path = "C:/Users/mayur/Downloads/ongc_train.csv"
traindata = pd.read_csv(path)
len(traindata)
```

```python
path = "C:/Users/mayur/Downloads/ongc_test.csv"
testdata = pd.read_csv(path)
testdata
```

```python
#predict the close price ]
#take the columns as ' close' from both train and test datasets
traindata = traindata.iloc[:,4:5].values
traindata
```

```python
testdata = testdata.iloc[:,4:5].values
```

```python
print(len(traindata)),print(len(testdata))
```

```python
#standardize the train and test datasets
minmax = MinMaxScaler()
train_scaled = minmax.fit_transform(traindata)
test_scaled = minmax.fit_transform(testdata)
```

```python
#verify the transformation
print(train_scaled)
```

```python
print(test_scaled)
```

```python
print(traindata[2000:])
```

```python
#set the lags(timestamp) for RNN
timestamp = 30
```

```python
#build the trainx and trainy datasets
#create the empty lists
trainx=[]; trainy=[]
```

```python
for i in range(timestamp,len(train_scaled)):
    trainx.append(train_scaled[i-timestamp:i,0])
    trainy.append(train_scaled[i,0])
    
```

```python
len(trainx[0])
```

```python
trainy[0]
```

```python
trainx[1]
```

```python
#change the trainx/trainy properties  to match the RNN format
trainx = np.array(trainx)
trainy = np.array(trainy)
trainx
```

```python
#reshape the trainx to convert it into a tensor
trainx = np.reshape(trainx,(trainx.shape[0],trainx.shape[1],1))
trainx.shape
```

```python
trainx[0]
```

```python
#build the network
reg = Sequential()
reg.add(LSTM(units=32,return_sequences=True,input_shape=(trainx.shape[1],1)))
```

```python
(trainx.shape[1],1)
```

```python
reg.add(Dropout(0.2))
```

```python
#LSTM Layer2
reg.add(LSTM(units=32,return_sequences=True))
reg.add(Dropout(0.2))
```

```python
#LSTM Layer3
reg.add(LSTM(units=32,return_sequences=True))
reg.add(Dropout(0.2))
```

```python
#LSTM Layer4 ->stock back
reg.add(LSTM(units=32))
reg.add(Dropout(0.2))
```

```python
#output layer
reg.add(Dense(units=1))
```

```python
#compile the network
reg.compile(optimizer='adam',loss='mean_squared_error')
```

```python
#train the model on the train data
reg.fit(trainx,trainy,epochs=20,batch_size=10)
```
